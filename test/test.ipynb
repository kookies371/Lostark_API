```python
class LostarkLoader():
    def __init__(self, JWT: str):
        self.JWT = JWT
    
    def load_card(self, character_name: str):
        response = requests.get(
            f"https://developer-lostark.game.onstove.com/armories/characters/{parse.quote(character_name)}/cards",
            headers = {
                'accpet': 'application/json',
                'authorization': f'bearer {JWT}'
            }
        )

        # TODO : check response == 200
        print(response)
        content = response.content
        card_json = json.loads(content)
        for card in card_json['Cards']:
            card['Tooltip'] = json.loads(card['Tooltip'])
        
        return card_json

Loader = LostarkLoader(JWT=JWT)
card = Loader.load_card('공깡짜장면')
pprint(card['Effects'])
print()
pprint(card['Effects'][0]['Items'])
print({'Description': '성속성 피해 +7.00%', 'Name': '세상을 구하는 빛 6세트 (18각성합계)'} in card['Effects'][0]['Items'])
```
